```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
dataset_file = 'C:/Users/ronak/Untitled Folder/Evaluation-dataset.csv' 
dataset = pd.read_csv(dataset_file, header=None, keep_default_na=False)
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
def determine_sentiment(text):
    if isinstance(text, str):
        sentiment_score = sid.polarity_scores(text)
        if sentiment_score['compound'] >= 0.05:
            return 'positive'
        elif sentiment_score['compound'] <= -0.05:
            return 'negative'
        else:
            return 'neutral'
    else:
        return 'neutral' 
```

```python
for index, row in dataset.iterrows():
    tire_review = row[0]
    garage_review = row[1]
    wait_time_review = row[2]
    
    # Determine sentiment for each review
    tire_sentiment = determine_sentiment(tire_review)
    garage_sentiment = determine_sentiment(garage_review)
    wait_time_sentiment = determine_sentiment(wait_time_review)
    
    # Print results
    print(f"Review {index + 1}:")
    print("Tire Review Sentiment:", tire_sentiment)
    print("Garage Review Sentiment:", garage_sentiment)
    print("Wait Time Review Sentiment:", wait_time_sentiment)
    print()
```
